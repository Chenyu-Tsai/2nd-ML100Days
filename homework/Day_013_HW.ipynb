```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
%matplotlib inline
from jupyterthemes import jtplot
jtplot.style()
import warnings
warnings.filterwarnings('ignore')
x=float('nan')

# 設定 data_path
dir_data = '../data/'
```

### 之前做過的處理

```python
# 讀取資料檔
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
app_train.shape
```

```python
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# 檢查每一個 column
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder, 以加入相關係數檢查
            app_train[col] = le.fit_transform(app_train[col])            
print(app_train.shape)
app_train.head()
```

```python
# 受雇日數為異常值的資料, 另外設一個欄位記錄, 並將異常的日數轉成空值 (np.nan)
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
app_train['DAYS_EMPLOYED'] = abs(app_train['DAYS_EMPLOYED'])
# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
```

## 練習時間
參考 Day 12 範例程式，離散化你覺得有興趣的欄位，並嘗試找出有趣的訊息

```python
work_data = app_train[['TARGET', 'DAYS_EMPLOYED']]
work_data['YEARS_EMPLOYED'] = abs(work_data['DAYS_EMPLOYED'] / 365)


bin_cut =  [0, 10, 20, 30, 40, 50]
work_data['YEARS_BINNED'] = pd.cut(work_data['YEARS_EMPLOYED'], bins = bin_cut)
work_data = work_data.dropna()

print(work_data['YEARS_BINNED'].value_counts())
work_data.head(10)
```

```python
year_group_sorted = np.sort(work_data['YEARS_BINNED'].unique())

plt.figure(figsize=(8,6), dpi = 300)
for i in range(len(year_group_sorted)):
    sns.distplot(work_data.loc[(work_data['YEARS_BINNED'] == year_group_sorted[i]) & \
                              (work_data['TARGET'] == 0), 'YEARS_EMPLOYED'], label = str(year_group_sorted[i]))
plt.title('KDE with Age groups other cases')
plt.savefig('other cases.png')
plt.show()
```

```python
plt.figure(figsize=(8,6), dpi = 300)
for i in range(len(year_group_sorted)):
    sns.distplot(work_data.loc[(work_data['YEARS_BINNED'] == year_group_sorted[i]) & \
                              (work_data['TARGET'] == 1), 'YEARS_EMPLOYED'], label = str(year_group_sorted[i]))
plt.title('KDE with Age groups late payment')
plt.savefig('sample.png')
plt.show()

```

```python
plt.figure(figsize = (8, 6), dpi = 300)

# 以年齡區間為 x, target 為 y 繪製 barplot
px = work_data['YEARS_BINNED']
py = work_data['TARGET']
sns.barplot(px, py)

# Plot labeling
plt.xticks(rotation = 75); plt.xlabel('Age Group (years)'); plt.ylabel('Failure to Repay (%)')
plt.title('Failure to Repay by Age Group');
plt.savefig('boxplot.png')
```
